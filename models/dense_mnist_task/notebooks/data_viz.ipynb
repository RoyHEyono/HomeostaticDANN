```python
import torch
import torchvision
from danns_eg.data.mnist import get_sparse_permutation_invariant_fashionmnist_dataloaders
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load FashionMNIST dataset
original_dataloader = get_sparse_permutation_invariant_fashionmnist_dataloaders(permutation_invariant=False)
dataloader = get_sparse_permutation_invariant_fashionmnist_dataloaders(permutation_invariant=False, brightness_factor=0.75)
```

```python
original_dataiter = iter(original_dataloader['train'])
original_images, _ = next(original_dataiter)
original_images = original_images.reshape(32,1,28,28)
```

```python
dataiter = iter(dataloader['train'])
images, labels = next(dataiter)
images = images.reshape(32,1,28,28)
```

```python
def imshow(img):
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()
```

```python
imshow(torchvision.utils.make_grid(original_images))
```

```python
imshow(torchvision.utils.make_grid(images))
```
